```python
def get_column_names(schemas, ds_name):
    column_details = schemas[ds_name]
    return [col['column_name'] for col in column_details]
schemas = json.load(open('data/retail_db/schemas.json'))
orders_columns = get_column_names(schemas, 'orders')
orders_columns
```

```python
orders = pd.read_csv(
    'data/retail_db/orders/part-00000',
    names=orders_columns
)
customers_columns = get_column_names(schemas, 'customers')
customers_columns
```

```python
customers = pd.read_csv(
    'data/retail_db/customers/part-00000',
    names=customers_columns
)
customers
```

```python
orders=orders.set_index('order_customer_id')
orders
```

```python
customers=customers.set_index('customer_id')
customers
```

```python
df1=customers.join(orders,how='inner').\
    reset_index('customer_id').\
    groupby('customer_id')['customer_id'].\
    agg(order_count='count')
df1.query('order_count > 10')
```

```python
import json
import pandas as pd
def get_column_names(schemas, ds_name, sorting_key='column_position'):
    column_details = schemas[ds_name]
    columns = sorted(column_details, key=lambda col: col[sorting_key])
    return [col['column_name'] for col in columns]
schemas = json.load(open('data/retail_db/schemas.json'))
orders_columns=get_column_names(schemas,'orders')
orders=pd.read_csv('data/retail_db/orders/part-00000',names=orders_columns)
orders

```

```python
orders.sort_values('order_customer_id',ascending=False)
```

```python
orders.sort_values(['order_date','order_customer_id'],ascending=[False,False])
```
